```python
tablero = [[" "]*3 for _ in range(3)]
jugador = "X"
maquina = "O"

tablero[1][1] = maquina

turno = 1

while True:
    print()
    print(tablero[0][0], "|", tablero[0][1], "|", tablero[0][2])
    print("---------")
    print(tablero[1][0], "|", tablero[1][1], "|", tablero[1][2])
    print("---------")
    print(tablero[2][0], "|", tablero[2][1], "|", tablero[2][2])
    print()

    print("Tu turno (X)")
    f = int(input("Fila (0-2): "))
    c = int(input("Columna (0-2): "))

    if tablero[f][c] != " ":
        print("Esa casilla ya est치 ocupada.")
        continue

    tablero[f][c] = jugador
    turno += 1

    if tablero[0][0] == tablero[0][1] == tablero[0][2] == jugador or \
       tablero[1][0] == tablero[1][1] == tablero[1][2] == jugador or \
       tablero[2][0] == tablero[2][1] == tablero[2][2] == jugador or \
       tablero[0][0] == tablero[1][0] == tablero[2][0] == jugador or \
       tablero[0][1] == tablero[1][1] == tablero[2][1] == jugador or \
       tablero[0][2] == tablero[1][2] == tablero[2][2] == jugador or \
       tablero[0][0] == tablero[1][1] == tablero[2][2] == jugador or \
       tablero[0][2] == tablero[1][1] == tablero[2][0] == jugador:
        print("Ganaste t칰... 쯖칩mo? 游땹")
        break

    print("Turno de la m치quina (O)")

    gano = False
    for i in range(3):
        if tablero[i].count(maquina) == 2 and tablero[i].count(" ") == 1:
            for j in range(3):
                if tablero[i][j] == " ":
                    tablero[i][j] = maquina
                    gano = True
                    break
        col = [tablero[0][i], tablero[1][i], tablero[2][i]]
        if col.count(maquina) == 2 and col.count(" ") == 1:
            for j in range(3):
                if tablero[j][i] == " ":
                    tablero[j][i] = maquina
                    gano = True
                    break
    if not gano:
        diag1 = [tablero[0][0], tablero[1][1], tablero[2][2]]
        if diag1.count(maquina) == 2 and diag1.count(" ") == 1:
            if tablero[0][0] == " ":
                tablero[0][0] = maquina
            elif tablero[1][1] == " ":
                tablero[1][1] = maquina
            else:
                tablero[2][2] = maquina
            gano = True

    if not gano:
        diag2 = [tablero[0][2], tablero[1][1], tablero[2][0]]
        if diag2.count(maquina) == 2 and diag2.count(" ") == 1:
            if tablero[0][2] == " ":
                tablero[0][2] = maquina
            elif tablero[1][1] == " ":
                tablero[1][1] = maquina
            else:
                tablero[2][0] = maquina
            gano = True

    if gano:
        print("La m치quina gana 游땙")
        break

    bloqueo = False
    for i in range(3):
        # Filas
        if tablero[i].count(jugador) == 2 and tablero[i].count(" ") == 1:
            for j in range(3):
                if tablero[i][j] == " ":
                    tablero[i][j] = maquina
                    bloqueo = True
                    break
        # Columnas
        col = [tablero[0][i], tablero[1][i], tablero[2][i]]
        if col.count(jugador) == 2 and col.count(" ") == 1:
            for j in range(3):
                if tablero[j][i] == " ":
                    tablero[j][i] = maquina
                    bloqueo = True
                    break
    # Diagonales
    if not bloqueo:
        diag1 = [tablero[0][0], tablero[1][1], tablero[2][2]]
        if diag1.count(jugador) == 2 and diag1.count(" ") == 1:
            if tablero[0][0] == " ":
                tablero[0][0] = maquina
            elif tablero[1][1] == " ":
                tablero[1][1] = maquina
            else:
                tablero[2][2] = maquina
            bloqueo = True

    if not bloqueo:
        diag2 = [tablero[0][2], tablero[1][1], tablero[2][0]]
        if diag2.count(jugador) == 2 and diag2.count(" ") == 1:
            if tablero[0][2] == " ":
                tablero[0][2] = maquina
            elif tablero[1][1] == " ":
                tablero[1][1] = maquina
            else:
                tablero[2][0] = maquina
            bloqueo = True

    # Si no hay nada que bloquear o ganar, pone la ficha donde pueda
    if not gano and not bloqueo:
        hecho = False
        for i in range(3):
            for j in range(3):
                if tablero[i][j] == " ":
                    tablero[i][j] = maquina
                    hecho = True
                    break
            if hecho:
                break

    # Verificar si tablero lleno (empate)
    lleno = True
    for fila in tablero:
        for casilla in fila:
            if casilla == " ":
                lleno = False
    if lleno:
        print("Empate. Nadie gana.游땛")
        break
```
